## Data visualization

### 1. Average speed by trip

```python
plt.figure(figsize = (7,7))
plt.title('average speed by trip')
sns.distplot(a = df['vitesse'], kde = False)
```

### 2.Percentage of trip by weekday

```python
# Calculation of % 
l= len(df[df['day_of_week']=='Monday'])/df.shape[0]*100
m= len(df[df['day_of_week']=='Tuesday'])/df.shape[0]*100
mer= len(df[df['day_of_week']=='Wednesday'])/df.shape[0]*100
j= len(df[df['day_of_week']=='Thursday'])/df.shape[0]*100
v= len(df[df['day_of_week']=='Friday'])/df.shape[0]*100
s= len(df[df['day_of_week']=='Saturday'])/df.shape[0]*100
d= len(df[df['day_of_week']=='Sunday'])/df.shape[0]*100

weekday_distribution= [l,m,mer,j,v,s,d]
print(weekday_distribution)
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize = (12,7))
plt.title('Number of trip by weekday in %')
sns.barplot(x= day_order, y= weekday_distribution,  order = day_order)
```

### 3.Percentage of trip by time slot

```python
time_slot1= len(df[df['time_slot']=='0-3h'])/df.shape[0]*100
time_slot2= len(df[df['time_slot']=='4-7h'])/df.shape[0]*100
time_slot3= len(df[df['time_slot']=='8-11h'])/df.shape[0]*100
time_slot4= len(df[df['time_slot']=='12-15h'])/df.shape[0]*100
time_slot5= len(df[df['time_slot']=='16-19h'])/df.shape[0]*100
time_slot6= len(df[df['time_slot']=='20-23h'])/df.shape[0]*100

time_slot_distribution= [time_slot1,time_slot2,time_slot3,time_slot4,time_slot5,time_slot6]
```

```python
time_slot_order = ['0-3h', '4-7h','8-11h','12-15h','16-19h','20-23h']
plt.figure(figsize = (12,7))
plt.title('Number of trip by time_slot in %')
sns.barplot(x=time_slot_order, y = time_slot_distribution)
```

### 4.Average trip distance by dayweek

```python
plt.figure(figsize = (12,7))
plt.title('Average trip distance by dayweek')
sns.barplot(x= 'day_of_week', y= 'distance', data = df, order = day_order)
```
